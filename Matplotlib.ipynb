```python
# 범례 lenend

data_dict = {'data_x':[1,2,3,4,5], 'data_y':[2,3,5,10,8]}   

plt.plot('data_x', 'data_y', data=data_dict, label = 'Price ($)')   # label 만 적을때는 안나옴 plt.legend()해줘야 나옴
plt.xlabel('X축', fontdict={'color':'b','size':'xx-large'})
plt.ylabel('Y축', fontdict={'color':'r','size':14})
plt.legend()
plt.show()

# 그래프 영역에 범례를 나타내려면 plot함수에 label 문자열을 지정하고 legend() 함수를 호출한다.
```

```python
data_dict = {'data_x':[1,2,3,4,5], 'data_y':[2,3,5,10,8]}   

plt.plot('data_x', 'data_y', data=data_dict, label = 'Price ($)')  
plt.xlabel('X축', fontdict={'color':'b','size':'xx-large'})
plt.ylabel('Y축', fontdict={'color':'r','size':14})
plt.legend(loc=0)
plt.show()
```

```python
data_1 = {'data_x':[1,2,3,4,5], 'data_y':[2,3,5,10,8]}   
data_2 = {'data_x':[1,2,3,4,5], 'data_y':[3,5,9,7,7]}

plt.plot('data_x', 'data_y', data=data_1, label = 'Price ($)')  
plt.plot('data_x', 'data_y', data=data_2, label = 'Demand ($)')  

plt.xlabel('X축', fontdict={'color':'b','size':'xx-large'})
plt.ylabel('Y축', fontdict={'color':'r','size':14})
plt.legend(loc=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
a = np.arange(0,2,0.2)

plt.plot(a,a,'r--',a,a**2,'bo',a,a**3,'g-')

plt.show()
```

```python
# 산점도 scatter plot
import numpy as np

np.random.seed(12)   # 일정하게 난수를 발생(숫자를 바꾸면 모양이 바뀌지만 그대로 고정된다)
n = 50

x = np.random.rand(n)
y = np.random.rand(n)

plt.scatter(x,y)
plt.show()
```

```python
import numpy as np

np.random.seed(5)   
n = 50

x = np.random.rand(n)
y = np.random.rand(n)
area = (30*np.random.rand(n)) ** 2
colors = np.random.rand(n)

plt.scatter(x,y, s=area, c=colors)
plt.show()
```

```python
import numpy as np

np.random.seed(5)   
n = 50

x = np.random.rand(n)
y = np.random.rand(n)
area = (30*np.random.rand(n)) ** 2
colors = np.random.rand(n)

plt.scatter(x,y, s=area, c=colors, alpha=0.4, cmap='Set1')
plt.colorbar()
plt.show()
```

```python
import numpy as np

np.random.seed(5)   
n = 50

x = np.random.rand(n)
y = np.random.rand(n)
z = np.random.rand(n)

area = (30*np.random.rand(n)) ** 2
colors = np.random.rand(n)

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x,y,z, s=area, c=colors, alpha=0.4, cmap='Set1')


plt.show()
```
